```python
#base code with added strong password validation
import PyPDF2
import re

# Function to validate password strength
def is_strong_password(password):
    """
    Checks if the password meets the following criteria:
    - Minimum 8 characters
    - At least one uppercase letter
    - At least one lowercase letter
    - At least one digit
    - At least one special character
    """
    return (
        len(password) >= 8 and
        re.search(r'[A-Z]', password) and
        re.search(r'[a-z]', password) and
        re.search(r'\d', password) and
        re.search(r'[!@#$%^&*(),.?":{}|<>]', password)
    )

# Function to create a password-protected PDF
def create_password_protected_pdf(input_pdf, output_pdf, password):
    # Check password strength before proceeding
    if not is_strong_password(password):
        print("Weak password. Use at least 8 characters, including uppercase, lowercase, number, and symbol.")
        return

    try:
        # Open the input PDF file
        with open(input_pdf, 'rb') as pdf_file:
            pdf_reader = PyPDF2.PdfReader(pdf_file)
            pdf_writer = PyPDF2.PdfWriter()

            # Copy all pages from the input PDF to the writer
            for page_num in range(len(pdf_reader.pages)):
                pdf_writer.add_page(pdf_reader.pages[page_num])

            # Encrypt the PDF with the provided password
            pdf_writer.encrypt(password)

            # Write the encrypted PDF to the output file
            with open(output_pdf, 'wb') as output_file:
                pdf_writer.write(output_file)

            print(f"Password-protected PDF saved as {output_pdf}")

    # Handle file not found error
    except FileNotFoundError:
        print(f"The file {input_pdf} was not found.")

    # Handle invalid PDF error
    except PyPDF2.errors.PdfReadError:
        print(f"The file {input_pdf} is not a valid PDF.")

    # Handle any other unexpected errors
    except Exception as e:
        print(f"Error: {e}")

# Collect user input for file names and password
input_pdf = input("Enter the name of the input PDF file (e.g., sample.pdf): ")
output_pdf = input("Enter the name for the output PDF file (e.g., protected_sample.pdf): ")
password = input("Enter the password to protect the PDF: ")

# Call the encryption function
create_password_protected_pdf(input_pdf, output_pdf, password)
```

```python
import os
import PyPDF2
import ipywidgets as widgets
from IPython.display import display, clear_output

# List all PDF files in the current directory
pdf_files = [f for f in os.listdir() if f.lower().endswith('.pdf')]

# Widgets for batch mode

# Allows selection of multiple PDF files
multi_select = widgets.SelectMultiple(
    options=pdf_files,
    description='Select PDFs:',
    style={'description_width': 'initial'},
    rows=6  # shows 6 files at once
)

# Password input for encrypting all selected PDFs
batch_password = widgets.Password(
    description='Password for all:',
    style={'description_width': 'initial'}
)

# Button to trigger batch encryption
batch_button = widgets.Button(
    description='Encrypt Selected',
    button_style='info'  # blue color
)

# Output area to display success or error messages
batch_status = widgets.Output()

# Batch encryption logic
def batch_encrypt(b):
    batch_status.clear_output()  # clear previous messages
    selected_files = multi_select.value  # get selected PDFs
    password = batch_password.value  # get entered password

    # Check if no PDFs are selected
    if not selected_files:
        with batch_status:
            print(" No PDFs selected.")
        return

    #  Check if password is empty
    if not password:
        with batch_status:
            print("Password is empty.")
        return

    #  Loop through each selected file
    for file in selected_files:
        output_name = f"protected_{file}"  # auto-generate output filename
        try:
            with open(file, 'rb') as f:
                reader = PyPDF2.PdfReader(f)

                #  Check if the PDF is already encrypted
                if reader.is_encrypted:
                    #  Attempt to decrypt using the provided password
                    reader.decrypt(password)  # <-- This line handles encrypted input

                writer = PyPDF2.PdfWriter()

                #  Copy all pages from input to output
                for page in reader.pages:
                    writer.add_page(page)

                # Encrypt with the provided password
                writer.encrypt(password)

                #  Save the encrypted PDF
                with open(output_name, 'wb') as out_file:
                    writer.write(out_file)

            #  Success message
            with batch_status:
                print(f" Encrypted: {output_name}")

        except Exception as e:
            #  Error message
            with batch_status:
                print(f" Error with {file}: {e}")

#  Connect button to encryption function
batch_button.on_click(batch_encrypt)

#  Display all widgets in the notebook
display(multi_select, batch_password, batch_button, batch_status)

"""if a pdf is already encryted and you try to encrypt it again with other multiple pdf's remember 
the password must be the same as of the earlier encryption otherwise that particular can't be decrypted and encrypted again
--> press CTRL and use the cursor to select multiple pdfs at once
"""
```

```python
import os
import PyPDF2
import ipywidgets as widgets
from IPython.display import display, clear_output
import re

# Step 1: List all PDF files in the current directory
pdf_files = [f for f in os.listdir() if f.lower().endswith('.pdf')]

# Step 2: Define default output folder as Downloads
default_downloads = os.path.join(os.path.expanduser("~"), "Downloads")

# Step 3: Create interactive widgets

# Widget to select multiple PDF files
multi_select = widgets.SelectMultiple(
    options=pdf_files,
    description='Select PDFs:',
    style={'description_width': 'initial'},
    rows=6  # Number of visible rows in the selection box
)

# Widget for entering the password used to decrypt existing encrypted PDFs
decrypt_password = widgets.Password(
    description='Decryption Password:',
    style={'description_width': 'initial'}
)

# Widget for entering the password to apply during encryption
encrypt_password = widgets.Password(
    description='Encryption Password:',
    style={'description_width': 'initial'}
)

# Widget to specify the output folder (default is Downloads)
output_folder = widgets.Text(
    value=default_downloads,
    description='Output Folder:',
    style={'description_width': 'initial'},
    placeholder='e.g., ./encrypted_pdfs'
)

# Informational label to guide the user about default behavior
default_path_info = widgets.Label(
    value="Note: If no folder path is provided, encrypted files will be saved to your Downloads folder."
)

# Button to clear all inputs and reset the interface
clear_button = widgets.Button(
    description='Clear All',
    button_style='warning'
)

# Button to start the encryption process
encrypt_button = widgets.Button(
    description='Encrypt Selected',
    button_style='success'
)

# Output area to display status messages and summary
status_output = widgets.Output()

# Step 4: Define a function to check password strength
def is_strong_password(pwd):
    """
    Validates password strength:
    - Minimum 8 characters
    - At least one uppercase letter
    - At least one lowercase letter
    - At least one digit
    - At least one special character
    """
    return (
        len(pwd) >= 8 and
        re.search(r'[A-Z]', pwd) and
        re.search(r'[a-z]', pwd) and
        re.search(r'\d', pwd) and
        re.search(r'[!@#$%^&*(),.?":{}|<>]', pwd)
    )

# Step 5: Define the batch encryption logic
def batch_encrypt(b):
    status_output.clear_output()  # Clear previous messages

    # Get user inputs
    selected_files = multi_select.value
    decrypt_pwd = decrypt_password.value
    encrypt_pwd = encrypt_password.value
    folder = output_folder.value.strip()

    # Use default Downloads folder if no path is provided
    if not folder:
        folder = default_downloads

    # Initialize counters for summary
    success_count = 0
    fail_count = 0

    # Validate inputs
    if not selected_files:
        with status_output:
            print("No PDFs selected.")
        return

    if not encrypt_pwd:
        with status_output:
            print("Encryption password is empty.")
        return

    if not is_strong_password(encrypt_pwd):
        with status_output:
            print("Weak password. Use at least 8 characters, including uppercase, lowercase, number, and symbol.")
        return

    # Create output folder if it doesn't exist
    if not os.path.exists(folder):
        os.makedirs(folder)

    # Loop through each selected file
    for file in selected_files:
        output_name = os.path.join(folder, f"protected_{file}")  # Auto-generate output filename
        try:
            with open(file, 'rb') as f:
                reader = PyPDF2.PdfReader(f)

                # If the PDF is encrypted, attempt to decrypt it
                if reader.is_encrypted:
                    reader.decrypt(decrypt_pwd)

                # Create a new PDF writer and copy all pages
                writer = PyPDF2.PdfWriter()
                for page in reader.pages:
                    writer.add_page(page)

                # Encrypt the new PDF with the provided password
                writer.encrypt(encrypt_pwd)

                # Save the encrypted PDF to the specified folder
                with open(output_name, 'wb') as out_file:
                    writer.write(out_file)

            # Increment success counter and show success message
            success_count += 1
            with status_output:
                print(f"Encrypted: {output_name}")

        except Exception as e:
            # Increment failure counter and show error message
            fail_count += 1
            with status_output:
                print(f"Error with {file}: {e}")

    # Display summary report
    with status_output:
        print(f"\nSummary: {success_count} succeeded, {fail_count} failed.")

# Step 6: Define the clear/reset function
def clear_all(b):
    """
    Resets all widgets to their default state.
    """
    multi_select.value = ()
    decrypt_password.value = ''
    encrypt_password.value = ''
    output_folder.value = default_downloads
    status_output.clear_output()

# Step 7: Connect buttons to their respective functions
encrypt_button.on_click(batch_encrypt)
clear_button.on_click(clear_all)

# Step 8: Display the full interface
display(
    multi_select,
    decrypt_password,
    encrypt_password,
    output_folder,
    default_path_info,
    widgets.HBox([encrypt_button, clear_button]),
    status_output
)
```
